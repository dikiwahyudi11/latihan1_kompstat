# No 01
Apakah ada dosen yang cenderung memberikan nilai cenderung lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya?

````python
# Code untuk Soal No 01 diberikan sebagai Contoh dan Bonus
# Pelajari dengan baik

nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen = nilai_dosen.set_index(['NIDN']) 
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
````

### Interpretasi No.1:

Terlihat bahwa terdapat perbedaan pemberian nilai di antara para dosen. 
Dosen yang cenderung memberikan nilai cenderung lebih tinggi adalah **Andri Basuki**, dengan rata rata nilai yang diberikan adalah 85.75. 
Sedangkan Dosen yang cenderung memberikan nilai cenderung lebih rendah adalah **Annisa Fitria**, dengan rata rata nilai yang diberikan adalah 60.09. 


# No 02
Menurut data yang ada, apakah terdapat perbedaan dalam pemberian nilai antara dosen dengan Pendidikan Terakhir S3 dan S2?

````python
dospend = pd.merge(dosen, nilai)
dosen_pend = dospend.groupby(dospend['Pendidikan_Terakhir']).mean()[['nilai']]
dosen_pend = dosen_pend.rename({'nilai':'mean_nilai'},axis=1)
stdscore = dospend.groupby(dospend['Pendidikan_Terakhir']).std()[['nilai']]
stdscore = stdscore.rename({'nilai':'std_nilai'}, axis=1)
dosen_pend = pd.concat([dosen_pend, stdscore], axis=1)
dosen_pend
````

Interpretasi No.2:

Terlihat adanya perbedaan dalam pemberian nilai antara dosen dengan pendidikan terakhir S3 dan S2. Dosen dengan pendidikan terakhir S3 cenderung memiliki rata-rata pemberian nilai yang lebih tinggi, yaitu sebesar 83.27, jika dibandingkan dengan dosen den pendidikan terakhir S2 dengan rata-rata pemberian nilai sebesar 64.17.

# No 03
Apakah terdapat perbedaan performa akademik antara mahasiswa yang berasal dari Madrasah/SMK dan SMU?

````python
mahasiswapend = pd.merge(mahasiswa,nilai)
mahasiswa_pend = mahasiswapend.groupby(mahasiswapend['Asal_Sekolah']).mean()[['nilai']]
mahasiswa_pend = mahasiswa_pend.rename({'nilai':'mean_nilai'},axis=1)
stdscore = mahasiswapend.groupby(mahasiswapend['Asal_Sekolah']).std()[['nilai']]
stdscore = stdscore.rename({'nilai':'std_nilai'}, axis=1)
mahasiswa_pend = pd.concat([mahasiswa_pend, stdscore], axis=1)
mahasiswa_pend
````

Interpretasi No.3:

Terlihat perbedaan dalam performa akademik antara mahasiswa yang berasal dari MA, SMK, dan SMU. Mahasiswa dengan pendidikan terakhir MA memiliki nilai rata-rata 70.78, sedangkan mahasiswa dengan pendidikan terakhir SMK memiliki nilai rata-rata 82.82, dan terakhir mahasiswa dengan pendidikan terakhir SMU memiliki nilai rata-rata 65.85.

# No 05
Apakah ada perbedaan prestasi akademik yang nampak di data antar angkatan mahasiswa?

````python
# Cara 4.2 dan 5 sama yaitu rata-rata kan dulu per NIM, bukan di-merge dulu kemudian dirata-ratakan
nilai_mahasiswa = nilai.groupby(nilai['nim']).mean()[['nilai']]
nilai_mahasiswa = nilai_mahasiswa.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_mahasiswa = pd.merge(mahasiswa, nilai_mahasiswa, left_on='nim', right_index=True)
nilai_mahasiswa = nilai_mahasiswa[['nim', 'Angkatan', 'mean_nilai']]
sb = nilai.groupby(nilai['nim']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_mahasiswa = nilai_mahasiswa.set_index(['nim']) 
nilai_mahasiswa = pd.concat([nilai_mahasiswa, sb], axis=1)

nilai_Angkatan = nilai_mahasiswa.groupby(nilai_mahasiswa['Angkatan']).mean()[['mean_nilai']]

sb = nilai_mahasiswa.groupby(nilai_mahasiswa['Angkatan']).std()[['mean_nilai']]
sb = sb.rename({'mean_nilai': 'std_nilai'}, axis=1) 
sb

nilai_AngkatanStd = pd.merge(nilai_Angkatan, sb, left_on='Angkatan', right_index=True)
nilai_AngkatanStd
````

Interpretasi 05:

Variabel nilai_mahasiswa untuk menggabungkan nim dan nilai, lalu dihitung mean_nilai dan std_nilai setiap mahasiswa, selanjutnya groupby Angkatan, dan didapatkan hasil seperti di atas.
Mean Nilai Angkatan 2019 dan 2020 tidak berbeda jauh, Standard deviasi masing-masing angkatan cukup besar, artinya masing-masing angkatan memiliki banyak variasi nilai. Dapat disimpulkan kedua angkatan tersebut tidak terdapat perbedaan prestasi akademik.
